Since the dataset has 54,000 entries, I will limit to 15 for ease of processing

```python
# randomly select 15 entries
rand_df = pd.DataFrame()
count = 0
while count <= 15:
  i = np.random.randint(0, len(df))
  new = df.loc[[i][:]]
  rand_df = pd.concat([rand_df, new], ignore_index = True)
  count += 1

rand_df
```

```python
# add entities and tags for each title to a new column
ents_tags_list = []
for title in rand_df['Title']:
  doc = nlp(title)
  ents_tags = ", ".join([f"{ent.text} {ent.label_}" for ent in doc.ents])
  ents_tags_list.append(ents_tags)
rand_df['Entities_Tags'] = ents_tags_list

rand_df
```

```python
# creating a combined view for ease of visualization
combined_titles = "\n".join(rand_df['Title'].tolist())

combined_doc = nlp(combined_titles)
# for ent in combined_doc.ents:
#   print(ent.text, ent.label)

displacy.render(combined_doc, style='ent', jupyter = True)
```

### Similarity Between Words

```python
!python -m spacy download en_core_web_md
# import en_core_web_md
nlp = spacy.load("en_core_web_md")
```

```python
# two random words, provided by generative AI
word1 = nlp('Serendipity')
word2 = nlp('Quixotic')

similarity_score = word1.similarity(word2)
print('The similarity score is: ', similarity_score)
```

### Similarity Between Phrases

```python
phrase1 = nlp("The vote to remove Altman showed how a philosophical movement devoted to the fear of A.I. had become an unavoidable part of tech culture")
phrase2 = nlp("Rosalynn Carter, first lady who championed mental health, dies at 96")

similarity_score = phrase1.similarity(phrase2)
print('The similarity score is: ', similarity_score)
```

### Similarity Between Docs

```python
# install necessary packages for web scraping
!pip install beautifulsoup4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests

url_list = ["https://www.washingtonpost.com/technology/2023/11/18/sam-altman-ilya-sutskever-openai/",
            "https://www.washingtonpost.com/obituaries/2023/11/19/rosalynn-carter-first-lady-jimmy-carter-dies/"]
doc_list = []

for url in url_list:
  response = requests.get(url)
  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    doc = soup.get_text()
    doc_list.append(doc)
  else:
    print(f"fail. status code: {response.status_code}")
```

```python
doc_list[0]
```

```python
doc_list[1]
```

The webscraping results aren't perfect and could use improvement especially at the beginning of the text, where it picks up headers), but for now they have pulled out a large portion of the text.

```python
doc1 = nlp(doc_list[0])
doc2 = nlp(doc_list[1])

similarity_score = doc1.similarity(doc2)
print('The similarity score is: ', similarity_score)
```
